# 3. Limpe o dataset, removendo linhas duplicadas e tratando valores faltantes

```python
dados = dados.drop_duplicates()
```

# 4. Utiliza o Matplotlib para visualizar as seguintes informações:

  a. Distribuição de acidez volátil por qualidade de vinho

  b.Distribuição do teor alcóolico por qualidade de vinho

  c.Relação entre acidez volátil e teor alcóolico

  d.Relalão entre teor alcóolico e qualidade do vinho

# a. Distribuição de acidez volátil por qualidade de vinho


```python
dados_qualidade = dados[['volatile acidity', 'quality']]
dados_qualidade.head(10)
```

```python
ax =sns.kdeplot(data=dados_qualidade, x='volatile acidity',y='quality',fill=True)
plt.suptitle('Distribuição de acidez volátil por qualidade de vinho')
plt.grid(alpha=0.2)
ax.plot()
```

```python
sns.histplot(data=dados_qualidade, element='bars', x='volatile acidity',y='quality')
plt.suptitle('Distribuição de acidez volátil por qualidade de vinho')
plt.grid(alpha=0.1)
ax.plot()
```

```python
dados_qualidade.plot(kind='scatter',x='volatile acidity',y='quality')
plt.suptitle('Distribuição de acidez volátil por qualidade de vinho')
plt.grid(alpha=0.1)
ax.plot()
```

# b.Distribuição do teor alcóolico por qualidade de vinho

```python
dados_qualidade_alcohol = dados[['alcohol','quality']]
```

```python
sns.kdeplot(data=dados_qualidade_alcohol, x='alcohol', y='quality', fill=True)
plt.suptitle('Distribuição de teor alcóolico por qualidade de vinho')
ax.plot()
```

```python
sns.histplot(data=dados_qualidade_alcohol, element='bars', x='alcohol',y='quality')
plt.suptitle('Distribuição do teor alcoólico por qualidade de vinho')
plt.grid(alpha=0.1)
ax.plot()
```

Observa-se que os melhores vinhos possuem um alto teor alcoólico. Os vinhos de qualidade 9 apresentam teor alcoólico variando entre os níveis 12 e 13. Quanto maior o teor alcoólico podemos visualizar uma tendência de aumento na qualidade. Contudo, a maioria dos vinhos possuem teor alcoólico  em torno de 9 e 10 e qualidade de nível 5 a 6. Há uma tendência de que o  aumento do teor alcoólico corrobore para um aumento na qualidade.

# c.Relação entre acidez volátil e teor alcóolico

```python
dados_acidez_alcohol = dados[['volatile acidity','alcohol']]
```

```python
ax =sns.scatterplot(data=dados_acidez_alcohol , x='volatile acidity', y='alcohol', linewidths=0.1)
plt.suptitle('Relação entre acidez volátil e teor alcoólico')
ax.grid(True, alpha=0.2)
ax.plot()
```

# d.Relação entre teor alcóolico e qualidade do vinho

```python
dados_qualidade_alcohol = dados[['alcohol','quality']]
```

```python
ax =sns.scatterplot(data=dados_qualidade_alcohol , x='alcohol',y='quality',linewidths=0.1)
plt.suptitle('Relação entre o teor alcoólico e qualidade de vinho')
ax.grid(True, alpha=0.2)
ax.plot()
```
