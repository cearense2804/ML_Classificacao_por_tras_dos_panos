# a. DistribuiÃ§Ã£o de acidez volÃ¡til por qualidade de vinho


```python
dados_qualidade = dados[['volatile acidity', 'quality']]
dados_qualidade.head(10)
```

```python
ax =sns.kdeplot(data=dados_qualidade, x='volatile acidity',y='quality',fill=True)
plt.suptitle('DistribuiÃ§Ã£o de acidez volÃ¡til por qualidade de vinho')
plt.grid(alpha=0.2)
ax.plot()
```

```python
sns.histplot(data=dados_qualidade, element='bars', x='volatile acidity',y='quality')
plt.suptitle('DistribuiÃ§Ã£o de acidez volÃ¡til por qualidade de vinho')
plt.grid(alpha=0.1)
ax.plot()
```

```python
dados_qualidade.plot(kind='scatter',x='volatile acidity',y='quality')
plt.suptitle('DistribuiÃ§Ã£o de acidez volÃ¡til por qualidade de vinho')
plt.grid(alpha=0.1)
ax.plot()
```

# b.DistribuiÃ§Ã£o do teor alcÃ³olico por qualidade de vinho

```python
dados_qualidade_alcohol = dados[['alcohol','quality']]
```

```python
sns.kdeplot(data=dados_qualidade_alcohol, x='alcohol', y='quality', fill=True)
plt.suptitle('DistribuiÃ§Ã£o de teor alcÃ³olico por qualidade de vinho')
ax.plot()
```

```python
sns.histplot(data=dados_qualidade_alcohol, element='bars', x='alcohol',y='quality')
plt.suptitle('DistribuiÃ§Ã£o do teor alcoÃ³lico por qualidade de vinho')
plt.grid(alpha=0.1)
ax.plot()
```

Observa-se que os melhores vinhos possuem um alto teor alcoÃ³lico. Os vinhos de qualidade 9 apresentam teor alcoÃ³lico variando entre os nÃ­veis 12 e 13. Quanto maior o teor alcoÃ³lico podemos visualizar uma tendÃªncia de aumento na qualidade. Contudo, a maioria dos vinhos possuem teor alcoÃ³lico  em torno de 9 e 10 e qualidade de nÃ­vel 5 a 6. HÃ¡ uma tendÃªncia de que o  aumento do teor alcoÃ³lico corrobore para um aumento na qualidade.
